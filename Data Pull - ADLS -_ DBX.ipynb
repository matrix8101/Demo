```python
configs = {
  "fs.azure.account.auth.type": "OAuth",
  "fs.azure.account.oauth.provider.type":
    "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",

  "fs.azure.account.oauth2.client.id":
    dbutils.secrets.get("adls-scope", "client-id"),

  "fs.azure.account.oauth2.client.secret":
    dbutils.secrets.get("adls-scope", "client-secret"),

  "fs.azure.account.oauth2.client.endpoint":
    f"https://login.microsoftonline.com/{dbutils.secrets.get('adls-scope', 'tenant-id')}/oauth2/token"
}
```

```python
dbutils.fs.mount(
  source="abfss://input@demoakshaysa.dfs.core.windows.net/",
  mount_point="/mnt/input",
  extra_configs=configs
)
```

```python
dbutils.fs.ls("/mnt/input")
```

```python
df_bronze = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/mnt/input/")
```
